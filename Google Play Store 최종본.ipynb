```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
```

```python
browser = webdriver.Chrome()
```

```python
search_word = input("검색하시고자 하는 앱 이름을 입력하세요.")

url = "https://play.google.com/store/search?q={}&c=apps"

final_url = url.format(search_word)
browser.get(final_url)

time.sleep(3)
browser.implicitly_wait(5)
app_url = browser.find_element(By.CLASS_NAME, 'Si6A0c').get_attribute('href')

browser.get(app_url)
time.sleep(5)
```

```python
browser.find_element(By.XPATH,'//*[@id="yDmH0d"]/c-wiz[2]/div/div/div[1]/div[2]/div/div[1]/c-wiz[4]/section/header/div/div[2]/button').click()
```

```python
#모달창 스크롤 내리기
#이건 3번만한거다. range(3)이니까

import pandas as pd

modal = browser.find_element(By.CLASS_NAME, 'fysCi')

data_list = []

for i in range(3):
    
    browser.execute_script('arguments[0].scrollTo(0, arguments[0].scrollHeight)', modal)
    
    time.sleep(5)
    
    containers = browser.find_elements(By.CLASS_NAME, 'RHo1pe')
    
    for container in containers:
    
        user_name = container.find_element(By.CLASS_NAME, 'X5PpBb').text
        
        rating = container.find_element(By.CLASS_NAME, 'iXRFPc').get_attribute('aria-label')[10]
        
        date = container.find_element(By.CLASS_NAME, 'bp9Aid').text
        
        content = container.find_element(By.CLASS_NAME, 'h3YV2d').text

        data = {
            "user_name": user_name,
            "rating": rating,
            "date": date,
            "content": content        
        }
        #이게 중요하다 스크롤을 내리면서 계속 겹치는게 생기니까 겹치는걸 방지하기 위함이다.
        if data not in data_list:
            data_list.append(data)
            print(len(data_list))

df = pd.DataFrame(data_list)
df
```

```python
#스크롤 끝까지 내리기

before_location = browser.execute_script("return document.body.scrollHeight")

while True:
    
    browser.execute_script("window.scrollTo(0,document.body.scrollHeight)")
        
    time.sleep(5)
    
    after_location = browser.execute_script("return document.body.scrollHeight")
    
    if before_location == after_location:
        break
        
    else:
        before_location = browser.execute_script("return document.body.scrollHeight")
```

```python
#지정한만큼 스크롤내리기

before_location = browser.execute_script("return window.pageYOffset")

while True:
        
    browser.execute_script("window.scrollTo(0,{})".format(before_location + 500))
    
    time.sleep(5)
    
    after_location = browser.execute_script("return window.pageYOffset")
    
    if before_location == after_location:
        break
        
    else:
        before_location = browser.execute_script("return window.pageYOffset")
```
